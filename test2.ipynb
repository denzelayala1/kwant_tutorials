```python
import kwant
import matplotlib.pyplot
import matplotlib.backends
```

```python
def billiard(pos):
    x, y= pos
    return x**2 + y**2 < 100^2

sys = kwant.Builder()
lat = kwant.lattice.square()
sys[lat.shape(billiard, (0,0))] = 4
sys[lat.neighbors()] = -1

sym_leed = kwant.TranslationalSymmetry([-1,0])
leed = kwant.Builder(sym_leed)
leed[(lat(0, y) for y in range(30, 60))] = 4
leed[lat.neighbors()] = -1
sys.attach_lead(leed)

sys = sys.finalized()
```

```python
kwant.plot(sys);
```

```python
ldos = kwant.ldos(sys, 0.2)
kwant.plotter.map(sys, ldos);
```
